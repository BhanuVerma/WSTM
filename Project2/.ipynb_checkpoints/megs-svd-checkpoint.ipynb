### All Imports and Constants

```python
import numpy as np
import os
from random import uniform, random, choice, sample
import time

from scipy import linalg
from scipy.sparse import csc_matrix

from sklearn.linear_model import orthogonal_mp

import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from PIL import Image,ImageOps
from pylab import *

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# Constants
types = ['Data/Person','Data/Sport']
_pixel = 255
no_of_images = 3
max_iter = 100

k_values = [int(max_iter/10), int(max_iter/3),int(max_iter/2),max_iter-1]
```

### Function to implement K-SVD

```python
def ksvd(Y, K, T, D=None, max_err=0, max_iter_val=10, approx=False):
    # Find the shape of the input
    (img_row_size, img_col_size) = Y.shape

    # init the dict
    if D is None:
        D = rand(img_row_size, K)

    # norm the dict
    for i in range(K):
        D[:,i] /= norm(D[:,i])
        
    X = zeros((K,img_row_size))
    err = inf
    iter_count = 0

    while iter_count < max_iter_val and err > max_err:
        X = orthogonal_mp(D, Y, T, max_err)
        atom_slots = range(K)
        np.random.shuffle(atom_slots)
        not_used_atoms = []

        for (i, j) in zip(atom_indices, range(K)):
            x_using = nonzero(X[i,:])[0]

            if len(x_using) == 0:
                not_used_atoms.append(i)
                continue

            X[i,x_using] = 0
            error_res = Y[:,x_using] - dot(D,X[:,x_using])

            try:
                U, s, Vt = svd(error_res)
            except LinAlgError:
                # This handles the 'SVD did not converge' issue
                return None, None
            
            D[:,i] = U[:,0]
            X[i,x_using] = s[0]*Vt.T[:,0]

        err_repres = Y - dot(D,X)
        err_norms = (norm(err_repres[:,n]) for n in range(img_col_size))
        err_indices = sorted(zip(err_norms, range(img_col_size)), reverse=True)

        for (not_used_ind, err_tuple) in zip(not_used_atoms, err_indices):
            (err, err_idx) = err_tuple

            d = Y[:,err_idx].copy()
            d -= mean(d)
            d /= norm(d)
            D[:,not_used_ind] = d

        err_norms = [norm(err_repres[:,n]) for n in range(img_col_size)]
        err = max(err_norms)
        iter_count += 1

    return D,X
```

### Function to plot image

```python
def plot_image(images,figname = 1):
    plt.figure(figname, figsize=(20, 5))
    plt.subplot(1,len(images),1)
    plt.title("Original Image")
    plt.imshow(images[0],cmap = plt.get_cmap('gray'))
    
    for idx, image in enumerate(images[1:]):
        plt.subplot(1,len(images),idx+2)
        title = "For K = %d, PSNR =  %0.2f " % (k_values[idx], calc_psnr(images[0],image))
        plt.title(title)
        plt.imshow(image,cmap = plt.get_cmap('gray'))
```

### Function to calculate the PSNR 

```python
def calc_psnr(initial, final):
    mse_value = np.mean((initial - final) ** 2)
    if mse_value == 0: return 100
    return 20 * math.log10(255.0 / math.sqrt(mse_value))
```

### Function to simplify K-SVD function call

```python
def image_ksvd(image_vector, intercept_value):
    return ksvd(image_vector, intercept_value, None, D=None, max_err=0, max_iter_val=10, approx=False)
```

### Function to perform KSVD for a particular category - For different values of K

```python
def ksvd_category(image_paths):
    
        count = 0
        while count < no_of_images:
            
            # Accumlates the initial and compressed images for a particular image
            images = []
            
            # Randomly choose one image, open it and store it as a vector
            image_path = np.random.choice(image_paths,1)
            img = Image.open(image_path[0])
            initial_image = np.array(img.resize(img.size).convert('L'))/255.0
            images.append(255.0*initial_image)
            
            # For different k values we perform compression for the image
            for k_value in k_values:
                D,X = image_ksvd(initial_image,k_value)
                if D==None or X==None:
                    break    
                else:   
                    images.append(255.0*dot(D,X))
                    
            if len(images) != len(k_values)+1:
                continue
                
            count +=1
            
            plot_image(images, count+1)
```
