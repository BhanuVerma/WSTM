```python
import tensorflow as tf
from matplotlib.pyplot import imshow
import numpy as np
import scipy as sp
import os, os.path
from PIL import Image
%matplotlib inline
```

```python
size = (256,256)
animalPath = 'Data/Animal'
fungusPath = 'Data/Fungus'
geoPath = 'Data/Geological Formation'
personPath = 'Data/Person'
plantPath = 'Data/Plant, flora, plant life'
sportPath = 'Data/Sport'

dirList=os.listdir(geoPath)
count  = 0

animalSize = len([name for name in os.listdir(animalPath)])
fungusSize = len([name for name in os.listdir(fungusPath)])
geoSize = len([name for name in os.listdir(geoPath)])
personSize = len([name for name in os.listdir(personPath)])
plantSize = len([name for name in os.listdir(plantPath)])
sportSize = len([name for name in os.listdir(sportPath)])

for fname in dirList:
    if count == 0:
        img = Image.open(animalPath+'/'+fname)
        img = img.resize(size,Image.ANTIALIAS)
        imshow(np.asarray(img))
        count += 1
    else:
        break
```
